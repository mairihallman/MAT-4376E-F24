```python
# dates = ['S2','Sch_Departure', 'Act_Departure','Departure_Date','Year']
# df[dates] = df[dates].apply(pd.to_datetime)

df['S2'] = pd.to_datetime(df['S2']).dt.to_period('min')
df['Sch_Departure'] = pd.to_datetime(df['Sch_Departure']).dt.to_period('min')
df['Act_Departure'] = pd.to_datetime(df['Act_Departure']).dt.to_period('min')

# Convert to day periods
df['Departure_Date'] = pd.to_datetime(df['Departure_Date']).dt.to_period('D')

# Convert to year periods
df['Year'] = pd.to_datetime(df['Year'], format='%Y').dt.to_period('Y')
```

```python
low_cardinality_int_cols = [col for col in df.select_dtypes(include=['int']).columns if df[col].nunique() < 10]
object_cols = df.select_dtypes(include=['object']).columns.tolist()
high_cardinality_int_cols = [col for col in df.select_dtypes(include=['int']).columns if df[col].nunique() >= 10]
continuous_cols = df.select_dtypes(include=['float']).columns.tolist()
period_cols = [col for col in df.columns if pd.api.types.is_period_dtype(df[col])]

dtypes = {
    "Column": [],
    "Type": [],
    "Domain": [],
    "Missing": []
}

for col in low_cardinality_int_cols + object_cols:
    unique_vals = df[col].dropna().unique()
    dtypes["Column"].append(col)
    dtypes["Type"].append("Low Cardinality Integer" if col in low_cardinality_int_cols else "Object")
    dtypes["Domain"].append(set(unique_vals))
    dtypes["Missing"].append(df[col].isna().sum())

for col in high_cardinality_int_cols + continuous_cols:
    min_val = df[col].min()
    max_val = df[col].max()
    dtypes["Column"].append(col)
    dtypes["Type"].append("High Cardinality Integer" if col in high_cardinality_int_cols else "Continuous")
    dtypes["Domain"].append([min_val, max_val])
    dtypes["Missing"].append(df[col].isna().sum())

for col in period_cols:
    min_val = df[col].min()
    max_val = df[col].max()
    dtypes["Column"].append(col)
    dtypes["Type"].append("Period")
    dtypes["Domain"].append(set([min_val]) if min_val == max_val else [min_val, max_val])
    dtypes["Missing"].append(df[col].isna().sum())

df_dtypes = pd.DataFrame(dtypes)
df_dtypes
```

```python
low_cardinality_int_cols = [col for col in df.select_dtypes(include=['int']).columns if df[col].nunique() < 3]
object_cols = df.select_dtypes(include=['object']).columns.tolist()
# for col in object_cols:
#     df[col] = df[col].astype('string')
high_cardinality_int_cols = [col for col in df.select_dtypes(include=['int']).columns if df[col].nunique() >= 3]
continuous_cols = df.select_dtypes(include=['float']).columns.tolist()
period_cols = [col for col in df.columns if pd.api.types.is_period_dtype(df[col])]

dtypes = {
    "Column": [],
    "Type": [],
    "Domain": [],
}

for col in low_cardinality_int_cols + object_cols:
    unique_vals = df[col].dropna().unique()
    dtypes["Column"].append(col)
    dtypes["Type"].append('String' if df[col].dtype == 'object' else 'Boolean')
    dtypes["Domain"].append(set(unique_vals))

for col in high_cardinality_int_cols + continuous_cols:
    min_val = df[col].min()
    max_val = df[col].max()
    dtypes["Column"].append(col)
    dtypes["Type"].append(f'{df[col].dtype}'.replace("int64","Integer").replace("float64","Float"))
    dtypes["Domain"].append([min_val, max_val])
   

for col in period_cols:
    min_val = df[col].min()
    max_val = df[col].max()
    dtypes["Column"].append(col)
    dtypes["Type"].append("DateTime")
    dtypes["Domain"].append(set([min_val]) if min_val == max_val else [min_val, max_val])

df_dtypes = pd.DataFrame(dtypes)
df_dtypes
```

```python
# Convert Period columns to datetime using .to_timestamp() instead of pd.to_datetime
s2_datetime = df['S2'].apply(lambda x: x.to_timestamp() if pd.notnull(x) else pd.NaT)
sch_departure_datetime = df['Sch_Departure'].apply(lambda x: x.to_timestamp() if pd.notnull(x) else pd.NaT)
act_departure_datetime = df['Act_Departure'].apply(lambda x: x.to_timestamp() if pd.notnull(x) else pd.NaT)

# Calculate time differences in minutes, handling NaNs by setting invalid differences to NaN
sch_act_diff = (act_departure_datetime - sch_departure_datetime).dt.total_seconds() / 60
s2_sch_diff = (sch_departure_datetime - s2_datetime).dt.total_seconds() / 60
s2_act_diff = (act_departure_datetime - s2_datetime).dt.total_seconds() / 60

# Create a temporary DataFrame with these calculated differences for analysis
temp_df = pd.DataFrame({
    'Sch_Act_Flag': df['Sch_Act_Flag'],
    'S2_Sch_Flag': df['S2_Sch_Flag'],
    'S2_Act_Flag': df['S2_Act_Flag'],
    'Sch_Act_Diff': sch_act_diff,
    'S2_Sch_Diff': s2_sch_diff,
    'S2_Act_Diff': s2_act_diff
})

# Filter out rows with NaN values in the difference columns before performing min and max calculations
temp_df = temp_df.dropna(subset=['Sch_Act_Diff', 'S2_Sch_Diff', 'S2_Act_Diff'])

# Calculate min and max of time differences by flag values
sch_act_diff_by_flag_min = temp_df.groupby('Sch_Act_Flag')['Sch_Act_Diff'].min()
sch_act_diff_by_flag_max = temp_df.groupby('Sch_Act_Flag')['Sch_Act_Diff'].max()

s2_sch_diff_by_flag_min = temp_df.groupby('S2_Sch_Flag')['S2_Sch_Diff'].min()
s2_sch_diff_by_flag_max = temp_df.groupby('S2_Sch_Flag')['S2_Sch_Diff'].max()

s2_act_diff_by_flag_min = temp_df.groupby('S2_Act_Flag')['S2_Act_Diff'].min()
s2_act_diff_by_flag_max = temp_df.groupby('S2_Act_Flag')['S2_Act_Diff'].max()

# Display results
print("Minimum actual departure - scheduled departure:\n", sch_act_diff_by_flag_min)
print("Maximum actual departure - scheduled departure:\n", sch_act_diff_by_flag_max)
print("\nMinimum scheduled departure - arrival at S2:\n", s2_sch_diff_by_flag_min)
print("Maximum scheduled departure - arrival at S2:\n", s2_sch_diff_by_flag_max)
print("\nMinimum actual departure - arrival at S2:\n", s2_act_diff_by_flag_min)
print("Maximum actual departure - arrival at S2:\n", s2_act_diff_by_flag_max)
```
